# Scraping Website

```python
url="https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3Drealme"
```

# Response

```python
response=requests.get(url)
```

```python
response
```

# parsing

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
Mobiles=soup.select("._4rR01T")
```

```python
Mobiles
```

```python
mobile_list=[]
```

```python
for item in Mobiles:
    mobile_list.append(item.text)
    mobile_list
```

```python
mobile_list
```

```python
prices=soup.select('._1_WHN1')
```

```python
prices
```
